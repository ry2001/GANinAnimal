```python
#  GET ANIMAL NAME LABES
import csv
import os
import json
import re

# Path to the CSV file containing the data
# Path of the image folder
csv_path = "pokemon.csv"
image_dir = "img/pkmn/"

# Initialize an empty list to hold the labels
labels = []
nonlabels = []
indexlabels =[]
incount = 0
outcount = 0

# Open the CSV file and iterate through its rows
with open(csv_path, "r") as f:
    fieldnames = ("Folder", "Index", "Type")
    reader = csv.DictReader(f, fieldnames)
    for row in reader:
        # Get the folder and type values from the current row
        id_parts = row['Folder'].split(';')
        row['Folder'] = id_parts[0]
        row['Index'] = id_parts[1]
        row['Type'] = id_parts[2]
        folder = row["Folder"]
        type_val = row["Type"]
        ind = row['Index']
        
        # Build the path to the folder containing the images
        folder_path = os.path.join(image_dir, folder)
        # Iterate through the files in the folder
        for filename in os.listdir(folder_path):
            if filename.endswith(".png"):
                # Build the full path to the image file
                image_path = os.path.join(folder_path, filename)
                if re.match(r'\d+\.png', filename):
                    integer = int(re.findall(r'\d+', filename)[0])
                    if str(integer)== str(ind):
                        label = [os.path.join(folder, filename), type_val]
                        labels.append(label)

                        incount += 1
                elif re.match(r'\d+_flip\.png', filename):
                    integer = int(re.findall(r'\d+', filename)[0])
                    if str(integer)== str(ind):
                        label = [os.path.join(folder, filename), type_val]
                        labels.append(label)   
                        
                        incount += 1
                elif re.match(r'shiny\d+\.png', filename):
                    integer = int(re.findall(r'\d+', filename)[0])
                    if str(integer)== str(ind):
                        label = [os.path.join(folder, filename), type_val]
                        labels.append(label)
   
                        incount += 1
                elif re.match(r'shiny\d+_flip\.png', filename):
                    integer = int(re.findall(r'\d+', filename)[0])
                    if str(integer)== str(ind):
                        label = [os.path.join(folder, filename), type_val]
                        labels.append(label)

                        incount += 1
                else:
                    outcount += 1

# Build the JSON object and write it to a file
json_obj = {"labels": labels}
print("json_obj", json_obj)
with open("labels.json", "w") as f:
    json.dump(json_obj, f)
```

```python
# SORTING THE POKEMON IMAGES INTO ANIMAL FOLDERS
import os
import json
import shutil

# Read the JSON file
with open('newlabels.json', 'r') as f:
    data = json.load(f)

# Create a new directory to store sorted images
if not os.path.exists('new_sorted_images'):
    os.makedirs('new_sorted_images')

# Loop through each label in the JSON file
for label in data['labels']:
    # Get the image path and animal name
    image_path = os.path.join('img/pkmn', label[0])
    animal_name = label[1]
#     print(image_path, animal_name) 
    # Create a new directory for the animal if it doesn't exist
    animal_dir = os.path.join('new_sorted_images', animal_name)
#     print(animal_dir) # eg. sorted_images/Frog
    if not os.path.exists(animal_dir):
        os.makedirs(animal_dir)

    # Copy the image to the animal directory
    image_name = os.path.basename(image_path) # eg . shiny1.png
    print(image_name)
    new_image_path = os.path.join(animal_dir, image_name)
    print( new_image_path)
    shutil.copyfile(image_path, new_image_path)
```

```python
# ZIPPING FILES
import zipfile
import os

# Name of the new ZIP file
zip_file_name = 'new_sorted_images.zip'

# Path of the folder to be zipped
folder_path = 'new_sorted_images'

# Create a ZipFile object with 'w' mode for writing
zip_obj = zipfile.ZipFile(zip_file_name, 'w', zipfile.ZIP_DEFLATED)

# Iterate over all the files in the folder
for foldername, subfolders, filenames in os.walk(folder_path):
    for filename in filenames:
        # Construct the full file path
        file_path = os.path.join(foldername, filename)
        # Add the file to the ZIP file
        zip_obj.write(file_path)

# Close the ZipFile object
zip_obj.close()

print(f"{zip_file_name} has been created!")
```
